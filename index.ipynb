```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
# from sklearn.preprocessing import MinMaxScaler
```

```python
# Reading the data

df = pd.read_csv('Mall_Customers.csv')
plt.scatter(df['CustomerID'], df['Spending Score (1-100)'])
```

### 1. Generating clusters dynamically

```python
mean_squared_errors = []
# From the total number of customers
k_range = range(1, 15)

for k in k_range:
    kMeansCluster = KMeans(n_clusters=k)
    kMeansCluster.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
    mean_squared_errors.append(kMeansCluster.inertia_ / 10000)
```

#### Drawwing the Elbow graph to find the optimum K

```python
plt.title('Elbow Graph')

plt.xlabel('K values')
plt.ylabel('Mean Squared Errors')

plt.plot(k_range, mean_squared_errors)
```

### Optimal value of K from the Elbow graph
The optimal value from the graph can be obtained from the elbow of a graph, which is found at 5. The 4 clusters are drawn in  scatter plot in the

```python
# Using kMeans clustering algorithm to form clusters
km = KMeans(n_clusters=5)

y_predicted = km.fit_predict(df[['Spending Score (1-100)', 'Annual Income (k$)']])
df['cluster'] = y_predicted
```

```python
# Plotting the scatter plot
df1, df2, df3, df4, df5 = df[df['cluster']==0], df[df['cluster']==1], df[df['cluster']==2], df[df['cluster']==3], df[df['cluster']==4]


plt.title('Customer group for a Retail store')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')


plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'], color='blue')
plt.scatter(df2['Annual Income (k$)'], df2['Spending Score (1-100)'],  color='red')
plt.scatter(df3['Annual Income (k$)'], df3['Spending Score (1-100)'], color='green')
plt.scatter(df4['Annual Income (k$)'], df4['Spending Score (1-100)'], color='yellow')
plt.scatter(df5['Annual Income (k$)'], df5['Spending Score (1-100)'], color='purple')
```

### Analysis 

The result from the scatter graph groups the customer into 5 groups that we can label as
1. Customers with lower annual income and below average spending habit
2. Customers with lower annual income and higher spending
3. Customers with average income and higher spending
4. Customers with higher income and high spending
5. Customers with higher income and minimal spending observed
