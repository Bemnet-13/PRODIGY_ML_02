```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
# from sklearn.preprocessing import MinMaxScaler
```

```python
# Reading the data

df = pd.read_csv('Mall_Customers.csv')
plt.scatter(df['CustomerID'], df['Spending Score (1-100)'])
```

```python
km = KMeans(n_clusters=5)

y_predicted = km.fit_predict(df[['CustomerID', 'Spending Score (1-100)']])
df['cluster'] = y_predicted
```

```python
df1, df2, df3, df4, df5 = df[df['cluster']==0], df[df['cluster']==1], df[df['cluster']==2], df[df['cluster']==3], df[df['cluster']==4]

plt.scatter(df1['CustomerID'], df1['Spending Score (1-100)'], color='blue')
plt.scatter(df2['CustomerID'], df2['Spending Score (1-100)'], color='red')
plt.scatter(df3['CustomerID'], df3['Spending Score (1-100)'], color='green')
plt.scatter(df4['CustomerID'], df4['Spending Score (1-100)'], color='yellow')
plt.scatter(df5['CustomerID'], df5['Spending Score (1-100)'], color='purple')
```

### 1. Generating clusters dynamically

```python
mean_squared_errors = []
# From the total number of customers
k_range = range(3, 10)

for k in k_range:
    kMeansCluster = KMeans(n_clusters=k)
    km.fit(df[['CustomerID', 'Spending Score (1-100)']])
    mean_squared_errors.append(km.inertia_ / 1000)

mean_squared_errors
```

#### Drawwing the Elbow graph to find the optimum K

```python
plt.title('Elbow Graph')

plt.xlabel('')
plt.ylabel('')

plt.plot(k_range, mean_squared_errors)
```
